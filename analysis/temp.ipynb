```python
import pandas as pd
oneMonth = pd.read_csv('one_month_demand_rate_order.csv')
oneDay = pd.read_csv('order.csv')
print(oneDay.columns)
print(oneMonth.columns)
```

```python
merged = pd.merge(left = oneMonth, right = oneDay,left_on='dwv_order_make_haikou_1.order_id', right_on='dwv_order_make_haikou_1.order_id')
print(merged.columns)
merged.head()
```

```python

merged['matching_prob_diff'] = merged['matching_prob_x'] - merged['matching_prob_y']
merged['ride_distance_diff'] = merged['ride_distance_x'] - merged['ride_distance_y']
merged['detour_distance_diff'] = merged['detour_distance_x'] - merged['detour_distance_y']
merged['shared_distance_diff'] = merged['shared_distance_x'] - merged['shared_distance_y']
merged['saved_distance_for_seeker_diff'] = merged['saved_distance_for_seeker_x'] - merged['saved_distance_for_seeker_y']
merged['saved_distance_for_taker_diff'] = merged['saved_distance_for_taker_x'] - merged['saved_distance_for_taker_y']
merged['lambda_diff'] = merged['lambda_x'] - merged['lambda_y']
merged.head()
```

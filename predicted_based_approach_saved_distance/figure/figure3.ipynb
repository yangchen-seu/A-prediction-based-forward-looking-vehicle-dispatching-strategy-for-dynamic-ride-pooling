```python

font = {'family' : 'Times New Roman', 'size'   : 16}
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

labels = ['100:10', '100:15', '100:25' ] # 

mb = [0.86,
0.94,
0.95
]
dta = [0.93,
0.96,
0.96
]

fl = [0.56,
0.85,
0.87
]

x = np.arange(3)
total_width, n = 0.8,5
width = total_width / n

plt.figure(figsize=(9,6))

rects2 = plt.bar(x - width, mb, width, fc='#FFBE7A')
rects3 = plt.bar(x , dta, width, fc='#FA7F6F')
rects4 = plt.bar(x +  width, fl, width, fc='#82B0D2')

# 
for i in x:
    plt.text(i - width ,mb[i],
             mb[i],
             ha='center', 
             va='bottom',fontdict=font
            )
    plt.text(i  ,dta[i],
             dta[i],
             ha='center', 
             va='bottom',fontdict=font
            )
    
    plt.text(i+ width ,fl[i],
             fl[i],
             ha='center', 
             va='bottom',fontdict=font
            )

# 
plt.ylabel('Response rate',fontdict=font)
plt.xlabel('Order-driver ratios', fontdict=font)

plt.xticks(x,labels,fontproperties = 'Times New Roman', size = 14)
plt.legend(['MB','DTA','FL'],prop=font)
plt.ylim((0,1.5))

plt.yticks(fontproperties = 'Times New Roman', size = 14)



plt.tight_layout() # 
plt.savefig('./response_rate.png', dpi = 200)
plt.show()

# In[] Average shared distance
labels = ['100:10', '100:15', '100:25' ] # 

mb = [2670,
      3119,
3475,
]
dta = [3715,
       3545,
3195
]

fl = [4960,
      4033,
4120
]

x = np.arange(3)
total_width, n = 0.8,5
width = total_width / n

plt.figure(figsize=(9,6))

rects2 = plt.bar(x - width, mb, width, fc='#FFBE7A')
rects3 = plt.bar(x , dta, width, fc='#FA7F6F')
rects4 = plt.bar(x +  width, fl, width, fc='#82B0D2')

# 
for i in x:
    plt.text(i - width ,mb[i],
             mb[i],
             ha='center', 
             va='bottom',fontdict=font
            )
    plt.text(i  ,dta[i],
             dta[i],
             ha='center', 
             va='bottom',fontdict=font
            )
    
    plt.text(i+ width ,fl[i],
             fl[i],
             ha='center', 
             va='bottom',fontdict=font
            )


# 
plt.ylabel('Average shared distance for passengers (m)',fontdict=font)
plt.xlabel('Order-driver ratios', fontdict=font)

plt.xticks(x,labels,fontproperties = 'Times New Roman', size = 14)
plt.yticks(fontproperties = 'Times New Roman', size = 14)
plt.ylim((0,7500))
plt.legend(['MB','DTA','FL'],prop=font)

plt.tight_layout() # 
plt.savefig('./Average shared distance for passengers.png', dpi = 200)
plt.show()



# In[] Total saved distance for drivers
labels = ['100:10', '100:15', '100:25' ] # 

mb = [81,
110,
114
]
fl = [127,
188,
206
]
dta = [109,
       140,
171
]


x = np.arange(3)
total_width, n = 0.8, 4
width = total_width / n

plt.figure(figsize=(9,6))

rects2 = plt.bar(x - width, mb, width, fc='#FFBE7A')
rects3 = plt.bar(x , dta, width, fc='#FA7F6F')
rects4 = plt.bar(x +  width, fl, width, fc='#82B0D2')

# 
for i in x:

    plt.text(i - width ,mb[i],
             mb[i],
             ha='center', 
             va='bottom',fontdict=font
            )
    plt.text(i  ,dta[i],
             dta[i],
             ha='center', 
             va='bottom',fontdict=font
            )
    
    plt.text(i+ width ,fl[i],
             fl[i],
             ha='center', 
             va='bottom',fontdict=font
            )


# 
plt.ylabel('Total saved distance for drivers (km)',fontdict=font)
plt.xlabel('Order-driver ratios', fontdict=font)

plt.xticks(x,labels,fontproperties = 'Times New Roman', size = 14)
plt.yticks(fontproperties = 'Times New Roman', size = 14)
plt.legend(['MB','DTA','FL'],prop=font)

plt.tight_layout() # 
plt.savefig('./Total saved distance for drivers.png', dpi = 200)
plt.show()

# In[] Average detour distance for passengers
labels = ['100:10', '100:15', '100:25' ] # 

mb = [992,
895,
842
]
fl = [671,
627,
596
]
dta = [842,
       746,
729
]


x = np.arange(3)
total_width, n = 0.8, 4
width = total_width / n

plt.figure(figsize=(9,6))

rects2 = plt.bar(x - width, mb, width, fc='#FFBE7A')
rects3 = plt.bar(x , dta, width, fc='#FA7F6F')
rects4 = plt.bar(x +  width, fl, width, fc='#82B0D2')

# 
for i in x:
    plt.text(i - width ,mb[i],
             mb[i],
             ha='center', 
             va='bottom',fontdict=font
            )
    plt.text(i  ,dta[i],
             dta[i],
             ha='center', 
             va='bottom',fontdict=font
            )
    
    plt.text(i+ width ,fl[i],
             fl[i],
             ha='center', 
             va='bottom',fontdict=font
            )


# 
plt.ylabel('Average detour distance for passengers (m)',fontdict=font)
plt.xlabel('Order-driver ratios', fontdict=font)

plt.xticks(x,labels,fontproperties = 'Times New Roman', size = 14)
plt.yticks(fontproperties = 'Times New Roman', size = 14)
plt.legend(['MB','DTA','FL'],prop=font)

plt.tight_layout() # 
plt.savefig('./Average detour distance for passengers.png', dpi = 200)
plt.show()
# In[] Average pick up time for passengers
labels = ['100:10', '100:15', '100:25' ] # 


mb = [254,
158,
142
]


fl = [135,
106,
84
]

dta = [217,
142,
126
]

x = np.arange(3)
total_width, n = 0.8, 4
width = total_width / n

plt.figure(figsize=(9,6))



rects2 = plt.bar(x - width, mb, width, fc='#FFBE7A')
rects3 = plt.bar(x , dta, width, fc='#FA7F6F')
rects4 = plt.bar(x +  width, fl, width, fc='#82B0D2')


# 
for i in x:
    plt.text(i - width ,mb[i],
             mb[i],
             ha='center', 
             va='bottom',fontdict=font
            )
    plt.text(i  ,dta[i],
             dta[i],
             ha='center', 
             va='bottom',fontdict=font
            )
    
    plt.text(i+ width ,fl[i],
             fl[i],
             ha='center', 
             va='bottom',fontdict=font
            )


# 
plt.ylabel('Average waiting time for passengers (s)',fontdict=font)
plt.xlabel('Order-driver ratios', fontdict=font)

plt.xticks(x,labels,fontproperties = 'Times New Roman', size = 14)
plt.yticks(fontproperties = 'Times New Roman', size = 14)
plt.legend(['Average response time of MB','Average response time of DTA','Average response time of FL'],prop=font)

plt.tight_layout() # 
plt.savefig('./Average waiting time for passengers.png', dpi = 200)
plt.show()
```

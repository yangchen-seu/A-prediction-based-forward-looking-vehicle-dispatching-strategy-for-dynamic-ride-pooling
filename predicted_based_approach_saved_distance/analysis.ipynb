```python
import pandas as pd
order_weight = pd.read_csv('output/order_weight.csv')
order_num = pd.read_csv('output/order_num.csv')

df = pd.merge(left= order_weight, right=order_num, left_on=['x_coord','y_coord'],right_on=['x_coord','y_coord'])
df.head()
df.columns
```

```python
import matplotlib.pyplot as plt

waiting_utility = df['waitingWeight']
demand = df['order_num']

plt.plot(demand,waiting_utility)
plt.xlabel('waitingWeight')
plt.ylabel('order_num')
plt.title('Waiting Utility vs Demand')
plt.show()

import numpy as np

correlation_matrix = np.corrcoef(waiting_utility, demand)
correlation_coefficient = correlation_matrix[0, 1]

print("Correlation coefficient:", correlation_coefficient)
```

```python
import pandas as pd
month_od = pd.read_csv('Month_order.csv')
od0801_month = month_od[month_od['day'] == 1]
OD = pd.read_csv('OD.csv')
print(len(od0801_month),len(OD))
```

```python
print(od0801_month['lambda'].sum())
print(OD['lambda'].sum())
merged = pd.merge(left= OD, right= od0801_month, left_on=['origin_id','destination_id'],right_on=['O_location','D_location'],how='left')
merged.head()
```

```python
def func(a,b):
    if a == b:
        return 0
    elif a> b:
        return 1
    else:
        return 2
merged['target'] = merged.apply(lambda row: func(row['lambda_x'], row['lambda_y']), axis=1)
merged.head()
print(len(merged))
tmp = merged.groupby(['target']).count().reset_index()
print(tmp)
```
